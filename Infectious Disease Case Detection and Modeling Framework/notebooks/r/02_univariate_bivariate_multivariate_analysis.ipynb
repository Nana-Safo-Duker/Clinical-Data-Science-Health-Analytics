# Univariate, Bivariate, and Multivariate Analysis

## Cancer Incidence Data Analysis (R)

```r
# Load libraries
library(ggplot2)
library(dplyr)
library(corrplot)
library(factoextra)
library(FactoMineR)
library(cluster)

source('../../scripts/r/data_loader.R')
source('../../scripts/r/univariate_analysis.R')
source('../../scripts/r/bivariate_analysis.R')
source('../../scripts/r/multivariate_analysis.R')
```

```r
# Load data
df <- load_data('../../data/incd.csv')
df_clean <- clean_data(df)
```

## Univariate Analysis

```r
# Univariate analysis
result <- univariate_analysis(df_clean, 'Incidence_Rate')
```

## Bivariate Analysis

```r
# Bivariate analysis
if ('Annual_Count' %in% colnames(df_clean)) {
  bivariate_analysis(df_clean, 'Incidence_Rate', 'Annual_Count')
}
if ('Trend' %in% colnames(df_clean)) {
  categorical_bivariate_analysis(df_clean, 'Trend', 'Incidence_Rate')
}
```

## Multivariate Analysis

```r
# Multivariate analysis
multivariate_analysis(df_clean)
cluster_analysis(df_clean, n_clusters = 3)
```
